```python

def raf2(coin,amount):
  dic={}
  amnt=[]
  zeros=[1]
  for i in range(amount+1):
    amnt.append(i)
    zeros.append(0)

  zeros.pop(-1)


  dic["a"]=amnt
  for i in coin:
    dic[i]=[]
  dic[0]=zeros


  for i in range(1,len(coin)):
    for j in range(len(amnt)):
      if coin[i]>amnt[j]:
        dic[coin[i]].append(dic[coin[i-1]][j])
        
      else:
        c=coin[i-1]+ dic[coin[i]][amnt[j]-coin[i]]
        dic[coin[i]].append(c)
        
  return dic
```

```python
coin=[0,1,3,5]
amount=8
dic=raf2(coin,amount)
for i in dic:
  print(i,dic[i])
```
