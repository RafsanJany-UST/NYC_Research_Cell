```python
import pandas as pd
import numpy as np
data = 'https://raw.githubusercontent.com/RafsanJany-44/Thesis_Project/main/All_DATA/without_sn_and_Epoch/EEG_HMC.csv'
dataset = pd.read_csv(data)
X = dataset.iloc[:, 1:]
Y = dataset.iloc[:, 0]

from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.30, random_state=7)
```

```python
model_logreg = KNeighborsClassifier(n_neighbors=3)
model_logreg.fit(X_train, Y_train)
model_logreg.predict(X_test)
accuracy_score(Y_test,model_logreg.predict(X_test))
```

#Testing

```python
import lime
from lime import lime_tabular

explainer = lime_tabular.LimeTabularExplainer(
    training_data=np.array(X_train),
    feature_names=list(dataset.iloc[:0,1:]),
    mode='classification'
)
```

```python
exp = explainer.explain_instance(X_test.iloc[8],
                                 model_logreg.predict_proba,
                                 num_features=6,
                                 top_labels=4)
```

```python
exp.show_in_notebook(show_table=True, show_all=False)
```

```python

import matplotlib.pyplot as plt
with plt.style.context("ggplot"):
    exp.as_pyplot_figure()
```
