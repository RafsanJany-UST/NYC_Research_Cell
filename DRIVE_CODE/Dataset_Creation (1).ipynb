# Temporal Spatial

```python


def Temporal_Spatial(df):
  Cadence_Right = list(df.iloc[44:,44])[0]
  Cadence_Left = list(df.iloc[44:,45])[0]

  Speed_Right=list(df.iloc[41:,44])[0]
  Speed_Left=list(df.iloc[41:,45])[0]

  Stride_Length_Right = list(df.iloc[38:,44])[0]
  Stride_Length_Left = list(df.iloc[38:,45])[0]


  Step_Length_Right = list(df.iloc[35:,46])[0]
  Step_Length_Left = list(df.iloc[35:,47])[0]


  Step_Time_Right = 60/Cadence_Right
  Step_Time_Left = 60/Cadence_Right

  Single_Support_Right = list(df.iloc[56:,46])[0]
  Single_Support_Left = list(df.iloc[56:,47])[0]

  IDS_Right = list(df.iloc[53:,46])[0]
  IDS_Left = list(df.iloc[53:,47])[0]

  TDS_Right=list(df.iloc[92:,46])[0]
  TDS_Left=list(df.iloc[92:,47])[0]

  Total_Support_Right = list(df.iloc[47:,46])[0]
  Total_Support_Left = list(df.iloc[47:,45])[0]

  Weight_Accept_Right = list(df.iloc[53:,46])[0]
  Weight_Accept_Left = list(df.iloc[53:,47])[0]

  Weight_Release_Right =  list(df.iloc[53:,47])[0]
  Weight_Release_Left = list(df.iloc[53:,46])[0]

  Stance_Right =  list(df.iloc[47:,46])[0]
  Stance_Left = list(df.iloc[47:,47])[0]

  Swing_Right = list(df.iloc[56:,47])[0]
  Swing_Left = list(df.iloc[56:,46])[0]

  return  [
    Cadence_Right,
    Cadence_Left,
    Speed_Right,
    Speed_Left,
    Stride_Length_Right,
    Stride_Length_Left,
    Step_Length_Right,
    Step_Length_Left,
    Step_Time_Right,
    Step_Time_Left,
    Single_Support_Right,
    Single_Support_Left,
    IDS_Right,
    IDS_Left,
    TDS_Right,
    TDS_Left,
    Total_Support_Right,
    Total_Support_Left,
    Weight_Accept_Right,
    Weight_Accept_Left,
    Weight_Release_Right,
    Weight_Release_Left,
    Stance_Right,
    Stance_Left,
    Swing_Right,
    Swing_Left]
```
