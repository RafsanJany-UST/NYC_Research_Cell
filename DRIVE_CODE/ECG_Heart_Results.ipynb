```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay

def Confusion(confusion_matrices, t):
    class_names = ['arrhythmic', 'healthy', 'ischemic']  # Define the class names

    # Create subplots
    num_matrices = len(confusion_matrices)
    cols = 3
    rows = (num_matrices + cols - 1) // cols  # Calculate rows needed

    fig, axs = plt.subplots(rows, cols, figsize=(15, 5 * rows))
    axs = axs.flatten()  # Flatten the array of axes

    for i, (index, cm_values) in enumerate(confusion_matrices):
        cm = np.array(cm_values).reshape(3, 3)  # Adjust based on the actual number of classes
        disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)  # Use class names
        disp.plot(ax=axs[i], values_format='d', cmap='Greens')  # Change 'viridis' to any colormap you prefer
        axs[i].set_title('Confusion Matrix for ' + t + " " + str(index))

    # Hide unused subplots
    for j in range(i + 1, len(axs)):
        axs[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
from plotly.subplots import make_subplots  # Import make_subplots for creating subplots
import numpy as np
import math

def Confusion_plotly(confusion_matrices, t):
    num_matrices = len(confusion_matrices)
    cols = 3
    rows = math.ceil(num_matrices / cols)  # Calculate rows needed

    # Create subplots
    fig = make_subplots(
        rows=rows, cols=cols,
        subplot_titles=[f'Confusion Matrix for {t} {index}' for index, _ in confusion_matrices],
        vertical_spacing=0.1,
        horizontal_spacing=0.1
    )

    # Add heatmaps for each confusion matrix
    for i, (index, cm_values) in enumerate(confusion_matrices):
        cm = np.array(cm_values).reshape(3, 3)  # Adjust based on actual number of classes
        z = cm[::-1]  # Flip the confusion matrix to match Plotly's heatmap layout

        heatmap = go.Heatmap(
            z=z,
            x=['Pred Class 0', 'Pred Class 1', 'Pred Class 2'],  # Change labels based on actual class names
            y=['True Class 2', 'True Class 1', 'True Class 0'],  # Flip the labels to match heatmap orientation
            colorscale='Greens',
            showscale=True,
            text=cm,  # Display values in the matrix
            texttemplate="%{text}",
            hoverinfo="text"
        )

        row = i // cols + 1
        col = i % cols + 1
        fig.add_trace(heatmap, row=row, col=col)

    # Update layout for better readability in a paper
    fig.update_layout(
        height=300 * rows, width=1000,
        title_text="Confusion Matrices",
        showlegend=False,
        margin=dict(t=40, b=40, l=40, r=40)
    )

    # Customize subplot titles font
    fig.update_annotations(font_size=12)

    # Display the plot
    fig.show()
```

```python
import plotly.graph_objects as go

# Data for the plot
accuracy_values = [0.9570860266685486,
 0.9327519536018372,
0.9090695803144557,
 0.9164574146270752,
 0.9030720591545105]

number_of_clients = [10, 20, 30, 40, 50]

# Create a bar plot with stripes
fig = go.Figure(go.Bar(
    x=number_of_clients,
    y=accuracy_values,
    marker_color='royalblue',  # Bar color
    width=5,  # Control the width of the bars
    marker_pattern_shape="/",  # Add stripes to the bars
))

# Add accuracy values (three decimal places) on top of the bars
for i, value in enumerate(accuracy_values):
    # Format to show three digits after the decimal point
    formatted_value = f"{value:.3f}"
    fig.add_annotation(
        x=number_of_clients[i],
        y=value,
        text=formatted_value,  # Use formatted value for display
        showarrow=False,
        yshift=10,  # Position the text slightly above the bars
        font=dict(size=12)
    )

# Customize the layout (use default background, remove title)
fig.update_layout(
    xaxis_title='Number of Clients',
    yaxis_title='Accuracy',
    width=800,
    height=500,
    xaxis=dict(showgrid=False),  # Remove x-axis grid lines
    yaxis=dict(
        showgrid=False,  # Remove y-axis grid lines
        range=[0.75, 0.98]  # Adjust the y-axis range to focus on the significant differences
    ),
)

# Show the plot
fig.show()
```
