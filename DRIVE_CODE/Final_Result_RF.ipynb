```python
import numpy as np
import pandas as pd

dataset=pd.read_csv("https://raw.githubusercontent.com/RafsanJany-44/Research-NREM-REM/main/dataset/REM_NREM.csv")

target="Sleep_Stage"

classes = np.array(sorted(list(set(dataset[target]))))
```

```python
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE

def balance(X_temp, y_temp):
  smote = SMOTE()
  X_temp, y_temp= smote.fit_resample(X_temp, y_temp)

  return pd.concat([pd.DataFrame(X_temp), pd.DataFrame(y_temp)], axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
dataset[target]=encoder.fit_transform(dataset[target])
```

```python

set(list(dataset[target]))
```

```python
X = dataset.loc[:, dataset.columns!=target]
y = dataset[target]
```

```python
dataset2=balance(X,y)
```

```python
X = dataset2.loc[:, dataset2.columns!=target]
y = dataset2[target]
```

```python
#Splitting dataset into training and testing dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
print(y_test.value_counts())
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_depth = RandomForestClassifier(max_depth=28,random_state=0)
rf_depth.fit(X_train, y_train)
y_pred=rf_depth.predict(X_test)


from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
print("Accurecy: ",accuracy_score(y_test, y_pred))
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(rf_depth, classes=classes)

cm.fit(X_train, y_train)
cm.score(X_test, y_test)
cm.show()
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
cv = RepeatedStratifiedKFold(n_splits=5, n_repeats=3, random_state=1)
Kfolds=20
Kfolds_range = range (2,Kfolds+1)
scores={}
loop_time=[]
scores_list = []
for k in Kfolds_range:
  cv = RepeatedStratifiedKFold(n_splits=3, n_repeats=k, random_state=1)
  l=list(cross_val_score(rf_depth,X, y,cv=cv))
  avg=sum(l)/len(l)
  scores[k] = avg
  scores_list.append(avg)
  print(str(k)+"/"+str(Kfolds)+" round completed.........................AVG Accurecy: "+str(avg))
```
