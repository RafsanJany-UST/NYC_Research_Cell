```python
import numpy as np
import pandas as pd
```

```python
data  = pd.read_csv("https://raw.githubusercontent.com/RafsanJany-44/Machine-School/main/datasets/EEG_HMC_Wake_NREM_REM_3000.csv")
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
data["Sleep Stage"]=encoder.fit_transform(data["Sleep Stage"])
```

```python
X = data.loc[:, data.columns != "Sleep Stage"]
y = data["Sleep Stage"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python

from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(random_state = 2021)
```

#Dictionary  of hyperparameter values to search

```python
search_space = {
    'n_estimators' : [100,200,500],
    'max_depth' : [3,6,9],
    'gamma':[0.01,0.1],
    "learning_rate":[0.001,0.01,0.1,1]
}
```

```python
import sklearn.metrics as k
print(*k.get_scorer_names(),sep='\n')
```

```python
from sklearn.model_selection import GridSearchCV

gs = GridSearchCV(estimator = xgb,
                  param_grid = search_space,
                  scoring=['r2','neg_mean_squared_error'],
                  refit = 'r2',
                  cv = 5,
                  verbose =4)
```

```python
gs.fit(X_train, y_train)
```

```python
print(gs.best_estimator_) 
```

```python
print(gs.best_params_)
```

```python
print(gs.best_score_)
```

```python
df = pd.DataFrame(gs.cv_results_)
df = df.sort_values("rank_test_r2")
df.to_csv("hello.csv",index = False)
```
