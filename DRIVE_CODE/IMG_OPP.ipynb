```python
from PIL import Image

def resize_and_compress(input_path, output_path, new_width, new_height, compression_level=9):
    # Open the image file
    original_image = Image.open(input_path)

    # Resize the image to the specified width and height
    resized_image = original_image.resize((new_width, new_height))

    # Save the resized and compressed image
    resized_image.save(output_path, format='PNG', compress_level=compression_level)

# Example usage:
input_image_path = '/content/photo.png'
output_image_path = 'output_photo.png'
new_width = 300
new_height = 300
compression_level = 5  # You can adjust this value (0-9) for different compression levels

resize_and_compress(input_image_path, output_image_path, new_width, new_height, compression_level)
```

```python
import tinify
import os

def compress_to_target_size(input_path, output_path, target_size_kb, max_iterations=10):
    # Set your Tinify API key
    tinify.key = 'your_tinify_api_key'

    for iteration in range(max_iterations):
        # Compress the image with the current compression parameters
        source = tinify.from_file(input_path)
        source.to_file(output_path)

        # Check the resulting file size
        current_size = os.path.getsize(output_path)

        # Adjust parameters for the next iteration
        if current_size > target_size_kb * 1024:
            # Decrease quality
            tinify.from_file(input_path).to_file(output_path, {"quality": 100 - iteration * 10})
        else:
            break

    print(f"Final size: {current_size / 1024:.2f} KB")

# Example usage:
input_image_path = 'output_photo.png'
output_image_path = 'output_image_compressed.png'
target_size_kb = 100

compress_to_target_size(input_image_path, output_image_path, target_size_kb)
```
