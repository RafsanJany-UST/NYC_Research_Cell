#Functions

```python
 
def get_contenct(patient_folder_adrs,d=0):
  if d == 0:
    folder_adrs = sorted(glob(patient_folder_adrs+"/*"))[0]
    type_of = "post"
  else:
    folder_adrs = sorted(glob(patient_folder_adrs+"/*"))[1]
    type_of = "pre"

  #print(patient_folder_adrs)
  #print(post_folder_adrs)
  #print(pre_folder_adrs)

  op_type = os.path.basename(os.path.normpath(folder_adrs))

  for i in sorted(glob(folder_adrs+"/*")):
    h = os.path.basename(os.path.normpath(i)).split("_")
    patient_id = h[0]
    if h[1] == "heartrate" and h[2] == "seconds":
      break
  return i,patient_id,type_of
```

##Automated

```python
p_id = []
state_list = []
nni_counter = []
nni_mean =[]
nni_min = []
nni_max = []

for i in sorted(glob(path+"/*")):


  add_post,id,state = get_contenct(i,0)
  df_post = pd.read_csv(add_post)
  results_post = td.nni_parameters(df_post["Value"])
  p_id.append(id)
  state_list.append(state)

  nni_counter.append(results_post[0])
  nni_mean.append(results_post[1])
  nni_min.append(results_post[2])
  nni_max.append(results_post[3])



  add_pre,id, state = get_contenct(i,1)
  df_pre = pd.read_csv(add_pre)
  results_pre = td.nni_parameters(df_pre["Value"])
  p_id.append(id)
  state_list.append(state)
  nni_counter.append(results_pre[0])
  nni_mean.append(results_pre[1])
  nni_min.append(results_pre[2])
  nni_max.append(results_pre[3])
```

```python
dic_main = {
    "P_id"        : p_id,
    "State"       : state_list,
    'nni_counter' : nni_counter,
    'nni_mean'    : nni_mean,
    'nni_min'     : nni_min,
    'nni_max'     : nni_max
}
```

```python
df = pd.DataFrame.from_dict(dic_main)
df.to_csv('fitabase_06_04_2023.csv')
df
```
