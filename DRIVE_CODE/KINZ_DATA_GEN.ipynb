# Defined Functions

```python


def sigle_file_data(file_path):

  data_dict = read_mat(file_path)


  basic_info = os.path.basename(file_path).split("_")
  p_id =  basic_info[0].split("-")[-2]+"-"+ basic_info[0].split("-")[-1]
  trail_num = basic_info[1]
  walking_speed = basic_info[2]


  kinz_arr = np.empty([0,111])


  for main_row in range(len(data_dict['t_uniform'])):

    if data_dict['t_uniform'][main_row] in data_dict['footfall_event_times']:
      footfall_event_times = [1]
    else:
      footfall_event_times = [0]

    joint_positions = []
    for i in range (32):
      x = data_dict["joint_positions"][main_row][0][i]
      y = data_dict["joint_positions"][main_row][1][i]
      z = data_dict["joint_positions"][main_row][2][i]
      joint_positions.append(x)
      joint_positions.append(y)
      joint_positions.append(z)

    accel_energy = data_dict["accel_energy"][main_row].tolist()


    comb_row = [p_id,trail_num,walking_speed ,data_dict['t_uniform'][main_row]]+ accel_energy + [data_dict["accel_energy_total"][main_row]]+ footfall_event_times + joint_positions+ [data_dict['left_foot_swing'][main_row],data_dict['pelvis_pos'][main_row], data_dict['right_foot_swing'][main_row]]

    kinz_arr = np.append(kinz_arr, [comb_row], axis=0)

  return kinz_arr

```

# Importing Lib and files

```python
from pymatreader import read_mat
import pandas as pd
import numpy as np
from glob import glob
from tqdm.notebook import tqdm

#data_path = '/home/iqh4001/Iqram_WCM/MADE Lab/Kinect Project/Dataset/Data/Cleaned Data'

data_path = "/content/drive/MyDrive/Iqram Sir/OneDrive_1_6-12-2024"



import os
all_file_path=[]

for i in sorted(glob(data_path+"/*")):
    for j in glob(i+"/*"):
        if 'TS' in os.path.basename(j).split("-"):
            all_file_path.append(j)
```
