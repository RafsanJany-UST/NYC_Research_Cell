```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the body joints and their connections
body_joints = ['HEAD', 'NOSE', 'EYE_LEFT', 'EAR_LEFT', 'EYE_RIGHT', 'EAR_RIGHT', 'HAND', 'PELVIS',
               'SPINE_NAVAL', 'SPINE_CHEST', 'NECK', 'CLAVICLE', 'SHOULDER', 'ELBOW', 'WRIST',
               'HANDTIP', 'THUMB', 'HIP', 'KNEE', 'ANKLE']

connections = [
    ('HEAD', 'NOSE'), ('NOSE', 'EYE_LEFT'), ('NOSE', 'EYE_RIGHT'), ('EYE_LEFT', 'EAR_LEFT'),
    ('EYE_RIGHT', 'EAR_RIGHT'), ('HEAD', 'NECK'), ('NECK', 'SPINE_NAVAL'), ('SPINE_NAVAL', 'SPINE_CHEST'),
    ('SPINE_CHEST', 'PELVIS'), ('PELVIS', 'HIP'), ('HIP', 'KNEE'), ('KNEE', 'ANKLE'), ('SHOULDER', 'ELBOW'),
    ('ELBOW', 'WRIST'), ('WRIST', 'HAND'), ('HAND', 'HANDTIP'), ('HAND', 'THUMB'), ('SHOULDER', 'CLAVICLE'),
    ('CLAVICLE', 'NECK')
]

# Load the dataset
file_path = '/content/new_dataset.csv'
data = pd.read_csv(file_path)

# Function to create a NetworkX graph from body joint coordinates
def create_nx_graph(row):
    G = nx.Graph()

    # Add nodes with coordinates as attributes
    for joint in body_joints:
      print(joint)
      print(row[f'{joint}_X'])
      print(row[f'{joint}_Y'])
      print(row[f'{joint}_Z'])
      G.add_node(joint, pos=(row[f'{joint}_X'], row[f'{joint}_Y'], row[f'{joint}_Z']))

    # Add edges based on predefined connections
    for connection in connections:
        G.add_edge(*connection)

    return G

# Create a NetworkX graph for the first row
sample_nx_graph = create_nx_graph(data.iloc[0])

# Extract positions for plotting
pos = nx.get_node_attributes(sample_nx_graph, 'pos')

# Function to plot a 3D graph
def plot_3d_graph(G, pos):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    # Draw nodes
    for node, (x, y, z) in pos.items():
        ax.scatter(x, y, z, s=100)
        ax.text(x, y, z, node, size=12, zorder=1, color='k')

    # Draw edges
    for edge in G.edges():
        x = [pos[edge[0]][0], pos[edge[1]][0]]
        y = [pos[edge[0]][1], pos[edge[1]][1]]
        z = [pos[edge[0]][2], pos[edge[1]][2]]
        ax.plot(x, y, z, c='black')

    plt.show()

# Plot the sample graph
plot_3d_graph(sample_nx_graph, pos)
```
