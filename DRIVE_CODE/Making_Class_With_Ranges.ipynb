```python


def sub_wise_splt(df,col,flage):
  total_count = len(set(df[col]))
  subjects=sorted(set(df[col]))

  train_axis = []
  test_axis = []

  count = 0

  for i in subjects:
    if count <= flage:
      train_axis.append(df.query(col+" == '"+i+"'"))
    else:
      test_axis.append(df.query(col+" == '"+i+"'"))
    count+=1

  return pd.concat(train_axis, axis=0),pd.concat(test_axis, axis=0)
```

```python
def target_filtering(df,target,thrashold):
  drop=[]
  for i in range(len(df[target])):
    if df[target][i] < thrashold:
      drop.append(i)
  df = df.drop(drop).reset_index(drop=True)
  return df
```

```python
from tqdm.notebook import tqdm


df = pd.read_csv("/content/Fitbit_feature_merged_version_2023.06.26_ST-II&ST-V_Pre.csv")

target = 'sd1'


print(df[target].min())
print(df[target].max())

for i in tqdm(range(len(df[target]))):

  #First condition
  if df[target][i] <= 5:
    df[target][i] = "class_1"

  #Second condition
  elif 6 <= df[target][i] <= 10:
    df[target][i] = "class_2"

  #Third condition
  elif 11 <= df[target][i] <= 15:
    df[target][i] = "class_3"

  #Fifth condition
  elif 16 <= df[target][i] <= 20:
    df[target][i] = "class_4"

  #Sixth condition
  elif 21 <= df[target][i] <= 25:
    df[target][i] = "class_5"

  #Seventh condition
  elif 26 <= df[target][i]:
    df[target][i] = "class_6"
```
