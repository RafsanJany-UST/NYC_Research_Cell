#Checking Null/NaN values

```python
primary_dataset.isnull().sum()
```

```python
print('Total number of null value:', primary_dataset.isnull().sum().sum())
print('Total number of columns:', primary_dataset.isna().any().sum())
```

#Filling nan values with a number

```python
df1 = primary_dataset.fillna(10)
```

```python
df1
```

```python
print('Total number of null value:', df1.isnull().sum().sum())
print('Total number of columns:', df1.isna().any().sum())
```

```python
primary_dataset
```

#Filling Null value with next value

```python
df2 = primary_dataset.fillna(method='bfill')
```

```python
df2
```

#Filling Null with previous value

```python
df3 = primary_dataset.fillna(method='pad')
```

```python
df3
```

#Fill Null with previous column value

```python
df5 = primary_dataset.fillna(method='pad',axis=1)
```

```python
print("Total number of row: ",df5.isnull().sum().sum())
print("Total number of columns: ",df5.isna().any().sum())
```

#Fill Null with next column value

```python
df6 = primary_dataset.fillna(method='bfill',axis=1)
```

```python
print("Total number of row: ",df6.isnull().sum().sum())
print("Total number of columns: ",df6.isna().any().sum())
```

#<font color='coral'>Fill Null for different columns with differents values</font>

```python
df7 = primary_dataset.fillna({'Delta 15 N (o/oo)': 999,
	                            'Delta 13 C (o/oo)':666})
```

#<font color='coral'>Filling Null for different columns with mean value of that column values</font>

```python
df8 = primary_dataset.fillna(value = primary_dataset['Delta 15 N (o/oo)'].mean())
```

```python
df9 = primary_dataset.fillna({'Delta 15 N (o/oo)': primary_dataset['Delta 15 N (o/oo)'].mean(),
                              'Delta 13 C (o/oo)': primary_dataset['Delta 13 C (o/oo)'].mean()}) #vvvvvvi
```
