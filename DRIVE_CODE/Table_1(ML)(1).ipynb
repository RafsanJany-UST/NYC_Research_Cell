#<font color = 'coral'>How Traditional Programming Works

Lets write a program to find <b>Celsius</b> to <b>Fahrenheit</b>.<br>
<br>
We know the equation is,<br>
<br>
F = Â°C x 1.8 + 32

```python

def celTofar(cel):
  far = 1.8*cel+32
  return far
```

```python
print("If temperature in Celsius : 25, then the Fahrenheit is: ",celTofar(25))
```

```python
cel_data = [200,90,45,255,130,35,45,29,277,56]
far_data = []


for i in cel_data:
  far_data.append(celTofar(i))
```

```python
far_data
```

```python
import numpy as np
import pandas as pd


data = {
    "Celsius":cel_data,
    "Fahrenheit":far_data
}


df = pd.DataFrame(data)
```

```python
df.head(-1)
```

#<font color='coral'>How Machine Learning Works</font>

```python
from sklearn.linear_model import LinearRegression
X = df["Celsius"].values.reshape(-1, 1)
Y = df["Fahrenheit"].values.reshape(-1, 1)
reg = LinearRegression().fit(X, Y)
```

```python
m = reg.coef_[0]
c = reg.intercept_
```

```python
print("Slop:",m)
print("Intercept:",c)
```

```python

x = 18


y = m*x + c

print(y)
```
