```python
import ipywidgets as widgets
```

```python
widgets.Checkbox(
    value=False,
    description='Check me',
    disabled=False
)
```

```python
new_keys=[]
keys=['key1','key2','key3','key4']
for i in keys:
  i=widgets.Checkbox(
    value=False,
    description=i,
    disabled=False,
    indent=False
    )
  display(i)
  new_keys.append(i)
  

```

```python
for i in new_keys:
  print(i.value)
     
```
