```python
st=input()
n=int(input())
v=['a', 'e','i', 'o', 'u','A','E','I','O','U']
st2=""
for i in range(len(st)):
  if(i%3==0):
    if st[i] in v:
      st2=st2+"YES"
    else:
      st2=st2+"NO"
  else:
    st2=st2+st[i]

print(st2)
```

```python
st=input()
v=['a', 'e','i', 'o', 'u','A','E','I','O','U']
Symbol = ["-", "+", "@", "$", "&"]
st2=""
count=0
flag=0
for i in st:
  if(i in v):
    if(count>=len(Symbol)):
      flag=1
      break
     
    st2=st2+Symbol[count]
    count+=1
  else:
    st2=st2+i

if flag==0:
  print(st2)
else:
  print('“Sorry, there are more than 5 vowels”')
```
