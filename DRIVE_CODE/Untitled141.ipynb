```python
for i in list(all_info.columns):
  if i != "subject_id" and i != "record_id2":
    hr_aee[i] = ''
```

```python
def all_info_back(p_id):
  li = []
  for i in list(all_info.columns):
    if i != "subject_id" and i != "record_id2":
      li.append((i,list(all_info.query("subject_id == '"+p_id+"'")[i])))
  return li
```

```python


hr_aee_index = 0


hr_aee_subject = hr_aee["Patient_Id"][hr_aee_index]
infos = all_info_back(hr_aee_subject)

for i in infos:
  hr_aee[i[0]][hr_aee_index] = i[1][0]
```
