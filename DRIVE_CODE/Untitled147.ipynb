```python
def cut_rod(p, n):
    """
    Compute maximum revenue
    :param p: prices list
    :param n: length of a rod
    :return:
    """
    if n == 0:
        return 0
    best_p = -1
    for i in range(n):
        best_p = max(best_p, p[i] + cut_rod(p, n - i - 1))
    return best_p


def cut_rod_solution(p, n, pre_best_cut):
    """
    Compute maximum revenue and the way to cut
    :param p: prices list
    :param n: length of a rod
    :param pre_best_cut: previous way to cut
    :return: tuple of (best price, best way to cut)
    """
    if n == 0:
        return 0, pre_best_cut
    best_price = -1
    cur_best_cut = None
    for i in range(n):
        candidate, best_cut_candidate = cut_rod_solution(p, n - i - 1, pre_best_cut + [i + 1])
        if candidate + p[i] > best_price:
            best_price = candidate + p[i]
            cur_best_cut = best_cut_candidate
    return best_price, cur_best_cut
```
