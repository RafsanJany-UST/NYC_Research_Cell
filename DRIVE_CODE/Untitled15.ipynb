```python
s=input()
l=len(s)
if(l<4):
  print(s)

elif l>=4:
  j=1
  s2=""
  if (l%2==0):
    m1=(l/2)
    m2=m1+1
    for i in s:
      if(j==m1 or j==m2):
        s2=s2+str(ord(i))+" "
      else:
        s2=s2+i+" "
      j+=1
  
  else:
    m1=int(l/2)+1
    m2=m1-1
    m3=m1+1
    for i in s:
      if(j==m1 or j==m2 or j==m3):
        s2=s2+str(ord(i))+" "
      else:
        s2=s2+i+" "
      j+=1

print(s2)
```
