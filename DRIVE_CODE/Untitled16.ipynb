```python
st=[]
for i in range(3):
  st.append(input())

for s in st:
  s2=''
  a=0
  b=0
  c=0
  for i in s:
    if(i=="A"):
      if(a<1):
        s2=s2+i
        a=1
      else:
        s2=s2+str(a)
      a+=1
      b=0
      c=0
    elif (i=="B"):
      if(b<1):
        s2=s2+i
        b=1
      else:
        s2=s2+str(b)
      b+=1
      a=0
      c=0
    else:
      if(c<1):
        s2=s2+i
        c=1
      else:
        s2=s2+str(c)
      c+=1
      b=0
      a=0
  print(s2)
```

```python
st1=input()
st=input()
count = 0
  
for i in range(len(st)-len(st1)+1):
    k=0
    while (k<len(st1)):
        if (st[i + k]!=st1[k]):
            break
        k=k+1
    if (k==len(st1)):
        count=count+1
        k=0
print(count)
```
