```python
#from lab

cel_data = [200,90,45,255,130,35,45,29,277,56]
far_data = [392.0, 194.0, 113.0, 491.0, 266.0, 95.0, 113.0, 84.2, 530.6, 132.8]
```

```python
#making dataset
import numpy as np
import pandas as pd


data = {
    "Cel": cel_data,
    "Far": far_data
}

df = pd.DataFrame(data)
```

```python
type(data)
```

```python
df
```

```python
df["Cel"]
```

```python
from sklearn.linear_model import  LinearRegression

X = df["Cel"].values.reshape(-1,1)
Y = df["Far"].values.reshape(-1,1)

reg = LinearRegression().fit(X,Y)
```

```python
m = reg.coef_[0]
c = reg.intercept_
```

```python
m,c
```
