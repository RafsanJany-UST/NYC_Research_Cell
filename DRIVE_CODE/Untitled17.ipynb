Write a Python program that takes a string as user input and prints a list 
containing only the characters appearing odd times on that particular 
string. You have to consider the characters as case insensitive which means 
that the uppercase and lowercase letters are considered the same.
[You are not allowed to use the built-in count() function for solving this 
problem. Also, you can not use the concepts of ‘Dictionary’ in this 
problem.]
------------------------------------------------------------------------------
Sample Input 1:
Alohomora

Sample Output 1:
['l', 'o', 'h', 'm', 'r']

Explanation 1:
Here, the characters 'a' ('A' and 'a' are considered the same) and 'o' 
appear 2 times which is even. So, the output list will not contain any of 
these characters.
------------------------------------------------------------------------------
Sample Input 2:
Avada Kedavra

Sample Output 2:
['a', ' ', 'k', 'e', 'r']

Explanation 2:
Here, the characters 'v' and 'd' appear 2 times which is even. So, the 
output list will not contain these characters. All other characters appear 
in the odd number of times.
------------------------------------------------------------------------------
Sample Input 3:
Mimblemimble

Sample Output 3:
[]

Explanation 3:
Here, all the characters appear twice. So, the final output list is empty.

```python
st="Avada Kedavra".lower()

talika1=[]
talika2=[]

for i in range(len(st)):
  count=1
  if (st[i] not in talika1):
    for j in range(i+1,len(st)):
      if st[i]==st[j]:
        count+=1
    talika1.append(st[i])
    talika2.append(count)

talika3=[]
for i in range(len(talika2)):
  if talika2[i]%2!=0:
    talika3.append(talika1[i])

print(talika3)
```
