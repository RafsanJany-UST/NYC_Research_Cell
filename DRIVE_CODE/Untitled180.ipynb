```python
import pandas as pd
import numpy as np

df = pd.read_csv("/content/MADELab_Fitabase_feature_matrix_S00-S70_2024-03-25_19-06PM.csv")
```

```python
def Search_Null(dataset):
  dic={}
  for col in list(dataset.columns):
    rows = []
    flage = 0
    for row in range(dataset.shape[0]):
      if str(dataset[col][row]) == "nan":
        rows.append(row)
        flage = 1
    if flage ==1:
      dic[col] = rows

  return dic


def Search_Some(dataset,some):
  dic={}
  for col in list(dataset.columns):
    rows = []
    flage = 0
    for row in range(dataset.shape[0]):
      if str(dataset[col][row]) == some:
        rows.append(row)
        flage = 1
    if flage ==1:
      dic[col] = rows

  return dic
```
