```python
import pandas as pd
import numpy as np


def Search_Null(dataset):
  dic={}
  for col in list(dataset.columns):
    rows = []
    flage = 0
    for row in range(dataset.shape[0]):
      if str(dataset[col][row]) == "nan":
        rows.append(row)
        flage = 1
    if flage ==1:
      dic[col] = rows

  return dic





```

```python

# dict all nan value



df = pd.read_csv("/content/MADELab_Fitabase_feature_matrix_S00-S70_2024-03-25_19-06PM.csv")


target_col = "Distance Walked:"



total_nan_dict = Search_Null(df)
original_list = list(df.index)





for column in total_nan_dict.keys():

  if column != target_col and str(df[column].dtypes) != "object":

    null_value_index = total_nan_dict[column]   #nan rows

    # Using filter() to remove nan value containing rows
    filtered_list = list(filter(lambda x: x not in null_value_index, original_list))


    rows_to_select = filtered_list

    selected_value_target_col = df.iloc[rows_to_select, df.columns.get_loc(target_col)].to_numpy().reshape(-1, 1)
    mean_value = np.nanmean(df[target_col])
    selected_value_target_col = np.where(np.isnan(selected_value_target_col), mean_value, selected_value_target_col)




    selected_value_nan_col = df.iloc[rows_to_select, df.columns.get_loc(column)].to_numpy().reshape(-1, 1)



    seleted_valu_fill_col = df.iloc[null_value_index, df.columns.get_loc(target_col)].to_numpy().reshape(-1, 1)
    mean_value = np.nanmean(df[target_col])
    seleted_valu_fill_col = np.where(np.isnan(seleted_valu_fill_col), mean_value, seleted_valu_fill_col)



    import pandas as pd
    from sklearn.linear_model import LinearRegression



    # Model selection and training
    model = LinearRegression()


    model.fit(selected_value_target_col, selected_value_nan_col)


    # Predicting and evaluating the model
    y_pred = list(model.predict(seleted_valu_fill_col).flatten())



    for i,row in enumerate(null_value_index):
      df[column][row] = y_pred[i]


```

```python
total_nan_dict = Search_Null(df)
```
