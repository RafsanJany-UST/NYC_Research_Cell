```python
from graphviz import Digraph

dot = Digraph()

dot.attr(rankdir='TB', fontsize='12', fontname='Arial', fontweight='bold')

dot.node('input', label='{InputLayer|input_shape=[(None, 32, 3)]}', shape='record')
dot.node('conv1', label='{Conv1D|filters=64, kernel_size=3, activation="relu"}', shape='record')
dot.node('bn1', label='{BatchNormalization}', shape='record')
dot.node('pool1', label='{MaxPooling1D|pool_size=2}', shape='record')

dot.node('conv2', label='{Conv1D|filters=128, kernel_size=3, activation="relu"}', shape='record')
dot.node('bn2', label='{BatchNormalization}', shape='record')
dot.node('pool2', label='{MaxPooling1D|pool_size=2}', shape='record')

dot.node('conv3', label='{Conv1D|filters=256, kernel_size=3, activation="relu"}', shape='record')
dot.node('bn3', label='{BatchNormalization}', shape='record')
dot.node('pool3', label='{MaxPooling1D|pool_size=2}', shape='record')

dot.node('lstm1', label='{LSTM|units=100, return_sequences=True}', shape='record')
dot.node('lstm2', label='{LSTM|units=50}', shape='record')

dot.node('dense1', label='{Dense|units=100, activation="relu"}', shape='record')
dot.node('dropout1', label='{Dropout|rate=0.5}', shape='record')
dot.node('dense2', label='{Dense|units=50, activation="relu"}', shape='record')
dot.node('dropout2', label='{Dropout|rate=0.5}', shape='record')

dot.node('output', label='{Dense|units=1, activation="sigmoid"}', shape='record')

dot.edge('input', 'conv1')
dot.edge('conv1', 'bn1')
dot.edge('bn1', 'pool1')
dot.edge('pool1', 'conv2')
dot.edge('conv2', 'bn2')
dot.edge('bn2', 'pool2')
dot.edge('pool2', 'conv3')
dot.edge('conv3', 'bn3')
dot.edge('bn3', 'pool3')
dot.edge('pool3', 'lstm1')
dot.edge('lstm1', 'lstm2')
dot.edge('lstm2', 'dense1')
dot.edge('dense1', 'dropout1')
dot.edge('dropout1', 'dense2')
dot.edge('dense2', 'dropout2')
dot.edge('dropout2', 'output')

dot.render('model_block_diagram', format='png', cleanup=True)
```
