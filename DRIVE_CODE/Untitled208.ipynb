```python
import pandas as pd
import numpy as np
from glob import glob
import os

path = "/content/drive/MyDrive/Iqram Sir/Fit-Bit_Data"

saving_path = "/content/"
```

```python
all_subject_folder = sorted(glob(path+"/*"))
```

```python
final = []

main_columns = ['Subject','Condition','TotalDistance', 'TrackerDistance', 'LoggedActivitiesDistance',
      'VeryActiveDistance', 'ModeratelyActiveDistance', 'LightActiveDistance',
      'SedentaryActiveDistance', 'VeryActiveMinutes', 'FairlyActiveMinutes',
      'LightlyActiveMinutes', 'SedentaryMinutes', 'Calories', 'Floors',
      'CaloriesBMR', 'MarginalCalories', 'RestingHeartRate',
      'TotalMinutesAsleep', 'TotalTimeInBed', 'TotalTimeAwake',
      'TotalMinutesLight', 'TotalMinutesDeep', 'TotalMinutesREM']



for subject_folder in all_subject_folder:

  post_folder = sorted(glob(subject_folder+"/*"))[0]
  pre_folder = sorted(glob(subject_folder+"/*"))[1]

  for f in glob(pre_folder+"/*.csv"):

    info = os.path.basename(f).split("_")

    if info[1] == "sleepStagesDay":

      df_sleep = pd.read_csv(f)

      temp_df_sleep = df_sleep[df_sleep["TotalSleepRecords"]!=0].iloc[:,2:]
      column_means_sleep = temp_df_sleep.mean().to_list()

      print("pre",info[0],info[1])

    if info[1] == "dailyActivity":
      df_activity = pd.read_csv(f)

      temp_df_activity = df_activity[df_activity["TotalSteps"]!=0].iloc[:,2:]
      column_means_activity = temp_df_activity.mean().to_list()

      print("pre",info[0],info[1])



  final.append([info[0]]+["pre"]+column_means_sleep+column_means_activity)




  for f in glob(post_folder+"/*.csv"):

    info = os.path.basename(f).split("_")

    if info[1] == "sleepStagesDay":
      df_sleep = pd.read_csv(f)
      temp_df_sleep = df_sleep[df_sleep["TotalSleepRecords"]!=0].iloc[:,2:]
      column_means_sleep = temp_df_sleep.mean().to_list()

      print("post",info[0],info[1])

    if info[1] == "dailyActivity":
      df_activity = pd.read_csv(f)
      temp_df_activity = df_activity[df_activity["TotalSteps"]!=0].iloc[:,2:]
      column_means_activity = temp_df_activity.mean().to_list()
      print("post",info[0],info[1])




  final.append([info[0]]+["post"]+column_means_sleep+column_means_activity)
  print("-_-_-_-_-_-_-_-_-")



df = pd.DataFrame(final, columns=main_columns)
df.to_csv(saving_path+'output.csv', index=False)
```
