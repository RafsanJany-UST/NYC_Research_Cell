```python
from tqdm import tqdm


deci_3digit = []
for i in range(100,1000):
    deci_3digit.append(str(i))



def palindrome(string):
    if string == string[::-1]:
        return True
    else:
        return False



all_palindromes = []

for i in tqdm(deci_3digit):
    for j in deci_3digit:
        dot = int(i)*int(j)
        if palindrome(str(dot)) == True:
            all_palindromes.append(dot)

print(max(all_palindromes))
```

```python
def is_prime(n):
  if n <= 1:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True


i= 2
prime_th= 0
while(True):
  if is_prime(i) == True:
    prime_th+=1
    print(i,prime_th)
  if prime_th == 10001:
    print(i)
    break
  i+=1

```
