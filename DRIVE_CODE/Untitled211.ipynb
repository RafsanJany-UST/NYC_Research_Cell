```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm
from glob import glob
import os


file_path = "/content/drive/MyDrive/Iqram Sir/Data"
saving_path = "/content/"
```

```python
all_columns = ['Subject_ID',
 'Condition',
 'Avarage_TotalSteps',
 'Avarage_TotalDistance',
 'Avarage_TrackerDistance',
 'Avarage_LoggedActivitiesDistance',
 'Avarage_VeryActiveDistance',
 'Avarage_ModeratelyActiveDistance',
 'Avarage_LightActiveDistance',
 'Avarage_SedentaryActiveDistance',
 'Avarage_VeryActiveMinutes',
 'Avarage_FairlyActiveMinutes',
 'Avarage_LightlyActiveMinutes',
 'Avarage_SedentaryMinutes',
 'Avarage_Calories',
 'Avarage_Floors',
 'Avarage_CaloriesBMR',
 'Avarage_MarginalCalories',
 'Avarage_RestingHeartRate',
 'Avarage_Calories',
 'Avarage_SedentaryMinutes',
 'Avarage_LightlyActiveMinutes',
 'Avarage_FairlyActiveMinutes',
 'Avarage_VeryActiveMinutes',
 'Avarage_SedentaryActiveDistance',
 'Avarage_LightActiveDistance',
 'Avarage_ModeratelyActiveDistance',
 'Avarage_VeryActiveDistance',
 'Avarage_StepTotal',
 'Avarage_TotalSleepRecords',
 'Avarage_TotalMinutesAsleep',
 'Avarage_TotalTimeInBed',
 'Avarage_TotalSleepRecords',
 'Avarage_TotalMinutesAsleep',
 'Avarage_TotalTimeInBed',
 'Avarage_TotalTimeAwake',
 'Avarage_TotalMinutesLight',
 'Avarage_TotalMinutesDeep',
 'Avarage_TotalMinutesREM']

all_data = []




for single_patient in tqdm(sorted(glob(file_path+"/*"))):
  print("--------------------------")
  print("Subject Processing: ",os.path.basename(single_patient))

  condition = sorted(glob(single_patient+"/*"))

  for c_j,c in enumerate(['post','pre']):
    print("Processing: ",c)
    df_dA = pd.read_csv(glob(condition[c_j]+"/*dailyActivity*")[0])
    df_dA['ActivityDate'] = pd.to_datetime(df_dA['ActivityDate']).dt.date


    df_dC = pd.read_csv(glob(condition[c_j]+"/*dailyCalories*")[0])
    df_dC['ActivityDay'] = pd.to_datetime(df_dC['ActivityDay']).dt.date


    df_dI = pd.read_csv(glob(condition[c_j]+"/*dailyIntensities*")[0])
    df_dI['ActivityDay'] = pd.to_datetime(df_dI['ActivityDay']).dt.date

    df_dS = pd.read_csv(glob(condition[c_j]+"/*dailySteps*")[0])
    df_dS['ActivityDay'] = pd.to_datetime(df_dS['ActivityDay']).dt.date

    df_sD = pd.read_csv(glob(condition[c_j]+"/*sleepDay*")[0])
    df_sD['SleepDay'] = pd.to_datetime(df_sD['SleepDay']).dt.date


    df_sSD = pd.read_csv(glob(condition[c_j]+"/*sleepStagesDay*")[0])
    df_sSD['SleepDay'] = pd.to_datetime(df_sSD['SleepDay']).dt.date


    all_means = []

    # dailyActivity section
    resting_heart_dates = []
    removed_index = []

    for i in range(df_dA.shape[0]):
      if np.isnan(df_dA.iloc[i,-1]):
        removed_index.append(i)
      else:
        resting_heart_dates.append(df_dA.iloc[i,0])

    df_dA.drop(index=removed_index, inplace=True)
    df_dA.drop(columns=['ActivityDate'], inplace=True)

    means = df_dA.mean()
    all_means = all_means + means.tolist()



    # dailyCalories section
    removed_index = []
    for i in range(df_dC.shape[0]):
      if df_dC.iloc[i,0] not in resting_heart_dates:
        removed_index.append(i)

    df_dC.drop(index=removed_index, inplace=True)
    df_dC.drop(columns=['ActivityDay'], inplace=True)

    means = df_dC.mean()
    all_means = all_means + means.tolist()




    # dailyIntensities section
    removed_index = []
    for i in range(df_dI.shape[0]):
      if df_dI.iloc[i,0] not in resting_heart_dates:
        removed_index.append(i)

    df_dI.drop(index=removed_index, inplace=True)
    df_dI.drop(columns=['ActivityDay'], inplace=True)

    means = df_dI.mean()
    all_means = all_means + means.tolist()



    # dailySteps section
    removed_index = []
    for i in range(df_dS.shape[0]):
      if df_dS.iloc[i,0] not in resting_heart_dates:
        removed_index.append(i)

    df_dS.drop(index=removed_index, inplace=True)
    df_dS.drop(columns=['ActivityDay'], inplace=True)

    means = df_dS.mean()
    all_means = all_means + means.tolist()


    # sleepDay section
    removed_index = []
    for i in range(df_sD.shape[0]):
      if df_sD.iloc[i,0] not in resting_heart_dates:
        removed_index.append(i)

    df_sD.drop(index=removed_index, inplace=True)
    df_sD.drop(columns=['SleepDay'], inplace=True)

    means = df_sD.mean()
    all_means = all_means + means.tolist()



    # sleepStagesDay section
    removed_index = []
    for i in range(df_sSD.shape[0]):
      if df_sSD.iloc[i,0] not in resting_heart_dates:
        removed_index.append(i)

    df_sSD.drop(index=removed_index, inplace=True)
    df_sSD.drop(columns=['SleepDay'], inplace=True)

    means = df_sSD.mean()
    all_means = all_means + means.tolist()


    all_data.append([os.path.basename(single_patient),c]+all_means)






df_final = pd.DataFrame(all_data, columns=all_columns)
df_final.to_csv(saving_path+"final_data.csv", index=False)
```
