```python
import plotly.graph_objs as go
import shap

def plot_scatter_shap(shap_values, feature_name, xtick_size=12, ytick_size=12, xlabel_size=14, ylabel_size=14, axis_linewidth=2):
    x_vals = shap_values.data[:, shap_values.feature_names.index(feature_name)]
    y_vals = shap_values[:, feature_name].values

    scatter = go.Scatter(
        x=x_vals,
        y=y_vals,
        mode='markers',
        marker=dict(
            symbol='triangle-up',
            color='#ff7cf3',
            size=12,
            line=dict(width=2, color='black')
        )
    )

    layout = go.Layout(
        xaxis=dict(
            title=go.layout.xaxis.Title(
                text=feature_name,
                font=dict(size=xlabel_size, color="black", family="Arial", weight="bold")
            ),
            tickfont=dict(size=xtick_size, color="black", weight="normal"),
            showgrid=True,
            linecolor='#f67474',
            linewidth=axis_linewidth,
            mirror=True  # Adds axis lines to both top and bottom
        ),
        yaxis=dict(
            title=go.layout.yaxis.Title(
                text="Feature SHAP values",#f"SHAP value for {feature_name}",
                font=dict(size=ylabel_size, color="black", family="Arial", weight="bold")
            ),
            tickfont=dict(size=ytick_size, color="black", weight="normal"),
            showgrid=True,
            linecolor='#f67474',
            linewidth=axis_linewidth,
            mirror=True  # Adds axis lines to both left and right
        ),
        margin=dict(l=40, r=40, t=20, b=40),
        #plot_bgcolor='white',
        height=600,
        width=800
    )

    fig = go.Figure(data=[scatter], layout=layout)
    fig.show()


features = [
    "sd_ratio_Moderate-High", "sampen_Moderate-High", "hr_max_Moderate-High", "aEE_MAX(J/min)",
    "Is patient on Beta Blockers", "fft_rel_vlf_Moderate-Low", "HRR_240", "sdnn_Moderate-High",
    "slope_HR/Calories", "Vigorous_Activity_ratio(%)", "dfa_alpha1_Moderate-Low", "Moderate-High_Activity_ratio(%)"
]



for i in features:
    plot_scatter_shap(shap_values2, i, xtick_size=24, ytick_size=24, xlabel_size=24, ylabel_size=24, axis_linewidth=2)
```
