```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'col_x': range(10, 110, 10),
    'col_y': [5, 15, 25, 35, 45, 55, 65, 75, 85, 95]
})

# Basic scatter plot with distribution
sns.jointplot(x='col_x', y='col_y', data=df, kind='scatter', marginal_kws=dict(bins=20, fill=True))

plt.show()
```
