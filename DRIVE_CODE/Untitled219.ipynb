```python
import matplotlib.pyplot as plt
import networkx as nx

# Define the prerequisites and dependencies
prerequisites = {
    "MAT116": ["MAT125", "MAT120"],
    "CSE115+L": ["CSE173"],
    "ENG102": ["ENG103"],
    "MAT120": ["MAT130", "PHY107+L", "EEE141+L"],
    "CSE173": ["CSE215+L", "CSE231+L"],
    "ENG103": ["BEN205"],
    "CSE215+L": ["CSE225+L"],
    "MAT130": ["MAT250", "PHY108+L"],
    "PHY107+L": ["PHY108+L", "EEE141+L"],
    "CSE225+L": ["CSE373", "CSE311+L"],
    "MAT250": ["MAT350", "MAT361"],
    "CSE231+L": ["CSE332"],
    "MAT350": ["CSE299"],
    "EEE141+L": ["EEE111+L"],
    "MAT361": ["CSE373"],
    "CSE332": ["CSE323"],
    "CSE323": ["CSE331+L"],
    "CSE311+L": ["CSE327+L"],
    "CSE327+L": ["CSE425"],
    "CSE499A**": ["CSE499B"],
}

# Create a directed graph
G = nx.DiGraph()

# Add standalone nodes
standalone_nodes = [
    "MAT125", "ENG111", "EEE111+L", "EEE452", "PHY108+L", "PHI104",
    "BEN205", "CEE10", "CSE299", "CSE373", "CSE331+L", "CSE425",
    "CSE499B", "CSE498", "CHE101+L", "HIS102", "HIS103", "BIO103+L",
    "POL101/104", "SOC101/ANT101/ENV203/GEO205", "ELECT1", "ELECT2",
    "ELECT3", "OPEN ELECT"
]

# Add nodes and edges to the graph
for node in standalone_nodes:
    G.add_node(node)

for course, dependencies in prerequisites.items():
    for dep in dependencies:
        G.add_edge(course, dep)

# Define a function for topological sort
def topological_sort(graph):
    visited = set()
    stack = []
    order = []

    # Helper function to perform DFS
    def dfs(course):
        if course in visited:
            return
        visited.add(course)
        for neighbor in graph.get(course, []):
            dfs(neighbor)
        stack.append(course)

    # Start DFS for each course
    for course in graph:
        if course not in visited:
            dfs(course)

    # Reverse the stack to get topological order
    return stack[::-1]

# Create a dictionary for graph representation (dependencies)
graph_dict = {}
for course, dependencies in prerequisites.items():
    graph_dict[course] = dependencies

# Add standalone courses to the graph dictionary
for node in standalone_nodes:
    if node not in graph_dict:
        graph_dict[node] = []

# Perform topological sort
top_order = topological_sort(graph_dict)



# Plot the graph as per your earlier code
# Define a color map for nodes
color_map = []
for node in G:
    if node.startswith("CSE"):
        color_map.append("lightgreen")
    elif node.startswith("MAT"):
        color_map.append("lightblue")
    elif node.startswith("ENG"):
        color_map.append("lightcoral")
    else:
        color_map.append("pink")

# Use a circular layout for node positions
pos = nx.circular_layout(G)

# Draw edges
plt.figure(figsize=(20, 20))
nx.draw_networkx_edges(G, pos, edge_color="gray")

# Draw nodes with the color map
nx.draw_networkx_nodes(
    G,
    pos,
    node_color=color_map,
    node_size=3000
)

# Draw node labels
nx.draw_networkx_labels(
    G,
    pos,
    labels={node: node for node in G.nodes()},
    font_size=8,
    font_color="black",
    font_weight="bold"
)

# Add a title
plt.title("Prerequisites and Dependencies Graph", fontsize=14)

# Save the graph as an image (optional)
plt.savefig("prerequisites_graph.png", dpi=300, bbox_inches="tight")

# Show the graph
plt.axis("off")
plt.show()


# Print the topological order of courses
print("Topological Order of Courses:")
for course in top_order:
    print(course)
```
