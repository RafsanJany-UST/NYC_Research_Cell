```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
duration = 1  # seconds
freq = 10  # Hz (signal frequency)

# Time vectors for different sampling rates
t_160 = np.linspace(0, duration, int(160 * duration), endpoint=False)
t_128 = np.linspace(0, duration, int(12 * duration), endpoint=False)

# Generate sine wave signals
signal_160 = np.sin(2 * np.pi * freq * t_160)
signal_128 = np.sin(2 * np.pi * freq * t_128)

# Create separate subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 6))

# Plot 160 Hz signal
axes[0].plot(t_160, signal_160, label="160 Hz Sampling", color="b", marker="o", markersize=3)
axes[0].set_title("Signal Sampled at 160 Hz")
axes[0].set_xlabel("Time (seconds)")
axes[0].set_ylabel("Amplitude")
axes[0].grid()
axes[0].legend()

# Plot 128 Hz signal
axes[1].plot(t_128, signal_128, label="128 Hz Sampling", color="r", linestyle="dashed", marker="x")
axes[1].set_title("Signal Sampled at 128 Hz")
axes[1].set_xlabel("Time (seconds)")
axes[1].set_ylabel("Amplitude")
axes[1].grid()
axes[1].legend()

# Adjust layout and show
plt.tight_layout()
plt.show()
```
