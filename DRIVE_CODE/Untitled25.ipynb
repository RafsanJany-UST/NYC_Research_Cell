```python
input_dir = '/kaggle/input/santander-customer-transaction-prediction/'
df_train = pd.read_csv("/content/N1_N2_N3_10f.csv")
df_train
```

```python
var_colums = [c for c in df_train.columns if c not in ['Sleep_Stage']]
X = df_train.loc[:, var_colums]
y = df_train.loc[:, 'Sleep_Stage']

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2)
X_train.shape, X_valid.shape, y_train.shape, y_valid.shape
```

```python
xgboost.XGBClassifier().get_params()
```

```python
model_xgboost = xgboost.XGBClassifier(learning_rate=0.1,
                                      max_depth=5,
                                      n_estimators=5000,
                                      subsample=0.5,
                                      colsample_bytree=0.5,
                                      eval_metric='auc',
                                      verbosity=1)

eval_set = [(X_valid, y_valid)]

model_xgboost.fit(X_train,
                  y_train,
                  early_stopping_rounds=10,
                  eval_set=eval_set,
                  verbose=True)
```
