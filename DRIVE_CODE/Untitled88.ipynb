```python
l = 4
sum = 10
lis = [3, 7, 1, 5]

new =[]


for i in range(l):
  if sum > lis[i]:
    new.append(sum-lis[i])
  else:
    new.append(1)


ans=[]


for i in range(l):
  ans.append(lis[i]-new[i])


print("lis",lis)
print("new",new)
print("ans",ans)



sec_index=None
for i in range(l):
  if ans[i]<0:
    sec_index=i+1
    dup = -1*ans[i]


print("dup",dup)
fst_index = None
for i in range(l):
  if ans[i] == dup:
    fst_index = i+1


print(fst_index,sec_index)
```

```python
def two_sum(arr, target):
    n = len(arr)
    seen = {}
    for i in range(n):
        if target - arr[i] in seen:
            return (seen[target - arr[i]] + 1, i + 1)
        seen[arr[i]] = i
    return "IMPOSSIBLE"

if _name_ == '_main_':
    n, s = map(int, input().split())
    arr = list(map(int, input().split()))
    result = two_sum(arr, s)
    print(*result)
```

```python
l = 5
sum = 10
lis = [3, 8, 1, 10, 0]

new =[]


for i in range(l):
  if sum > lis[i]:
    new.append(sum-lis[i])
  else:
    new.append(None)

ans=[]

for i in range(l):
  if new[i] != None:
    ans.append(new[i]-lis[i])
  else:
    ans.append(None)


sec_index=None
for i in range(l):
  if ans[i] != None and ans[i]<0:
    sec_index=i+1
    dup = -1*ans[i]


fst_index = None
for i in range(l):
  if ans[i] != None and ans[i] == dup:
    fst_index = i+1

print("lis",lis)
print("new",new)
print("ans",ans)

print(fst_index,sec_index)
```
