```python
!pip install imbalanced-learn

from imblearn.over_sampling import SMOTE


def balance(x_temp, y_temp):
  smote = SMOTE()

  x_temp, y_temp = smote.fit_resample(x_temp, y_temp)

  return x_temp, y_temp
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Thesis_Data/REM_NREM.csv")

target = "Sleep_Stage"

```

#Label Encoding

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

df[target] = encoder.fit_transform(df[target])
```

```python
sorted(list(set(df[target])))
```

```python
df.columns
```

#Dataset Spliting

```python
x = df.loc[:,df.columns != target]          # df.loc[starting_row: ending_row, list_of_columns]

y = df[target]
```

```python
x.head()
```

```python
y.head()
```

#Dataset spliting for training and testing

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
df[target].value_counts()
```

```python
y_train.value_counts()
```

```python
x_train, y_train = balance(x_train, y_train)
```

```python
y_train.value_counts()
```

#Training Phase

##Random Forest

```python
%%time
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()

rf.fit(x_train,y_train)         #training line
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print("Accuracy: ",accuracy_score(y_test,y_pred))
```

#Model Save and Load

```python
import joblib

def save_model(model, model_name):
  joblib.dump(model,model_name)

def load_model(model_name):
  return joblib.load(model_name)
```

```python
save_model(rf,"RandomForest")
```
