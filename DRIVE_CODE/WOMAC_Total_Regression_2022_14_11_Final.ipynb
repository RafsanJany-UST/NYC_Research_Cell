#Decision Tree Regressor



```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor


for i in range(2,1000):
  dt_classifier = DecisionTreeRegressor(max_depth=i)
  dt_classifier.fit(X_new, y_new)
  y_pred = dt_classifier.predict(X_new)
  from sklearn.metrics import r2_score,mean_squared_error
  print("Training r2: ",r2_score(y_new,y_pred),", When Max Depth:",i)
  if r2_score(y_new,y_pred) ==1:
    mx=i-1
    break
```

```python
dt_classifier = DecisionTreeRegressor(max_depth=mx)
dt_classifier.fit(X_new, y_new)

y_pred = dt_classifier.predict(X_new)

pred_y_df = pred_dataframe(y_new,y_pred)
reg_plot(pred_y_df,width=15,height=10)
```

```python
explain_shap(dt_classifier,X_new,20)
```

#Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression


count=[]
mx_depth=[]

for i in range(2,150):
  rfr_classifier = RandomForestRegressor(max_depth=i)

  rfr_classifier.fit(X_new, y_new)
  y_pred = rfr_classifier.predict(X_new)
  r2=r2_score(y_new,y_pred)
  print("Training r2: ",r2,", When Max Depth:",i)
  count.append(r2)
  mx_depth.append(i)
  
print("Max R2: ",max(count)," Max Depth: ",mx_depth[count.index(max(count))])
```

```python
rfr_classifier = RandomForestRegressor(max_depth=mx_depth[count.index(max(count))])
rfr_classifier.fit(X_new, y_new)

y_pred = rfr_classifier.predict(X_new)

pred_y_df = pred_dataframe(y_new,y_pred)
reg_plot(pred_y_df,width=15,height=10)
```

```python
explain_shap(rfr_classifier,X_new,20)
```
