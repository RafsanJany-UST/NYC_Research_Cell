```python
# problem 5.1


def mode(lyst):
    x=lyst.copy()
    a=0
    b=''
    for i in x:
        f=x.count(i)
        if f>a:
            a=f
            b=str(i)
            x.remove(i)
        elif f==a:
            b+=', '+str(i)
    return b

def median(lyst):
    notun=sorted(lyst)
    if len(notun)%2==0:
        sum_of_middle_two_number=notun[len(notun)//2-1]+notun[len(notun)//2]
        moddhok=sum_of_middle_two_number/2
    else:
        moddhok=notun[len(notun)//2+1]
    return moddhok
    
def mean(lyst):
    somosti=sum(lyst)
    gor=somosti/len(lyst)
    return gor
    

if __name__ == "__main__":

    lyst = [3,1,7,1,4,10]
    print('List: ',lyst)
    print("Mode:",mode(lyst))
    print("Median:",median(lyst))
    print("Mean:",mean(lyst))
```
