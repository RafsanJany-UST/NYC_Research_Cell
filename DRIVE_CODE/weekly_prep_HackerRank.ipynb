```python
def timeConversion(s):
  
  if s[-2]=="A":
    if s[0:2]=="12":
      s=s.replace("AM","")
      s=s.split(":")
      return ("00"+":"+s[1]+":"+s[2])
    else:
      s=s.replace("AM","")
      s=s.split(":")
      return (s[0]+":"+s[1]+":"+s[2])
    
  else:
    if s[0:2]=="12":
      s=s.replace("PM","")
      s=s.split(":")
      return (s[0]+":"+s[1]+":"+s[2])
    s=s.replace("PM","")
    s=s.split(":")
    h=int(s[0])
    if h<12:
      h=12+h
    elif h==12:
      h="00"
    return (str(h)+":"+s[1]+":"+s[2])


print(timeConversion("12:45:54PM"))
```
