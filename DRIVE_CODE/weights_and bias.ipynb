```python
cel_in = [-40, -10, 0, 8, 15, 22, 38]
```

```python
def cel_to_far(c):
  f = c*1.8 + 32
  return f
```

```python
far_out = []
for i in cel_in:
   far_out.append(cel_to_far(i))
```

```python
far_out
```

#Neural Networks

```python
import numpy as np
```

```python
cel = np.array(cel_in,dtype=float)
far = np.array(far_out, dtype = float)
```

```python
import tensorflow as tf

#make a layer
l0 = tf.keras.layers.Dense(units = 1, input_shape=[1]) # 1 unit = 1 neurons

#define a model
model = tf.keras.Sequential([l0])
```

```python
model.compile(loss="mean_squared_error", 
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
%%time
history = model.fit(cel,far, epochs=500, verbose=False)
```

```python
print(model.predict([100.0]))
```

```python
print(cel_to_far(100))
```

```python
print("These are the layer variable, weights:", l0.get_weights()[0],", bias:",l0.get_weights()[1])
```

```python

#layer 1
l0 = tf.keras.layers.Dense(units = 3, input_shape=[1])
#hidden layer
l1 = tf.keras.layers.Dense(units = 2)
#output layer
l2 = tf.keras.layers.Dense(units = 1)

#define model
model = tf.keras.Sequential([l0,l1,l2])
```

```python
model.compile(loss="mean_squared_error", 
              optimizer=tf.keras.optimizers.Adam(0.1))
```

```python
%%time
model.fit(cel,far,epochs=500,verbose=False)
```

```python
print(model.predict([100.0]))
```

```python
print(cel_to_far(100))
```

```python
print("These are the layer variable, weights:", l0.get_weights()[0],", bias:",l0.get_weights()[1])
```

```python
print("These are the layer variable, weights:", l1.get_weights()[0],", bias:",l1.get_weights()[1])
```
