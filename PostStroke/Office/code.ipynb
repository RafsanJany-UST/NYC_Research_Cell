```python
import os
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, random_split
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load and preprocess the data
file_path = 'Final_Merged_Data.csv'
df = pd.read_csv(file_path)

# Define input and target columns
X_cols = [
    'Pside_AnkleAngles', 'Pside_KneeAngles', 'Pside_HipAngles', 'Pside_PelvisAngles',
    'Pside_GASnorm', 'Pside_RFnorm', 'Pside_VLnorm', 'Pside_BFnorm', 'Pside_STnorm', 'Pside_TAnorm', 'Pside_ERSnorm',
    'Nside_AnkleAngles', 'Nside_KneeAngles', 'Nside_HipAngles', 'Nside_PelvisAngles',
    'Nside_GASnorm', 'Nside_RFnorm', 'Nside_VLnorm', 'Nside_BFnorm', 'Nside_STnorm', 'Nside_TAnorm', 'Nside_ERSnorm'
]

y_col = 'Time post stroke (days)'

# Extract inputs and targets (Drop NaNs before scaling)
df.dropna(inplace=True)
X = df[X_cols].values
y = df[y_col].values

# Verify no NaNs after dropping
if np.isnan(X).any() or np.isnan(y).any():
    raise ValueError('NaNs still present in data after dropping!')
if np.isinf(X).any() or np.isinf(y).any():
    raise ValueError('Infinity values present in data after dropping!')
X = df[X_cols].values
subjects = df['Subject_ID'].values
y = df[y_col].values

# Scale inputs (Ensure proper alignment)
from sklearn.preprocessing import RobustScaler
scaler_X = StandardScaler()
scaler_y = MinMaxScaler(feature_range=(0, 1))

df.reset_index(drop=True, inplace=True)

# Extract inputs and targets
X = df[X_cols].values
y = df[y_col].values

# Print target value stats before scaling
print(f"Target (y) - Min: {np.min(y)}, Max: {np.max(y)}, Mean: {np.mean(y)}")

# Scale features and targets
X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

# Print target value stats after scaling
import matplotlib.pyplot as plt
plt.hist(y, bins=50, color='skyblue', edgecolor='black')
plt.title('Scaled Target (y) Distribution')
plt.xlabel('Scaled Time Post Stroke (days)')
plt.ylabel('Frequency')
plt.show()
print(f"Scaled Target (y) - Min: {np.min(y)}, Max: {np.max(y)}, Mean: {np.mean(y)}")

# Verify scaling integrity
if np.isnan(y).any():
    raise ValueError('NaN values found in the target variable after scaling.')
if np.isinf(y).any():
    raise ValueError('Infinite values found in the target variable after scaling.')
if np.var(y) == 0:
    raise ValueError('Zero variance found in the target variable after scaling.')
from sklearn.preprocessing import RobustScaler
scaler_X = StandardScaler()
scaler_y = MinMaxScaler(feature_range=(0, 1))

# Extract inputs and targets after resetting index to avoid alignment issues
df.reset_index(drop=True, inplace=True)
X = df[X_cols].values
y = df[y_col].values

# Scale features and targets
X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

# Verify scaling integrity
if np.isnan(y).any():
    raise ValueError('NaN values found in the target variable after scaling.')
if np.isinf(y).any():
    raise ValueError('Infinite values found in the target variable after scaling.')
if np.var(y) == 0:
    raise ValueError('Zero variance found in the target variable after scaling.')
from sklearn.preprocessing import RobustScaler
scaler_X = StandardScaler()
scaler_y = MinMaxScaler(feature_range=(0, 1))
X = scaler_X.fit_transform(X)
y = scaler_y.fit_transform(y.reshape(-1, 1)).flatten()

# Check for NaN and zero variance in targets
if np.isnan(y).any():
    raise ValueError('NaN values found in the target variable after scaling.')
if np.isinf(y).any():
    raise ValueError('Infinite values found in the target variable after scaling.')
if np.var(y) == 0:
    raise ValueError('Zero variance found in the target variable after scaling.')
if np.isnan(y).any():
    raise ValueError('NaN values found in the target variable after scaling.')
if np.isinf(y).any():
    raise ValueError('Infinite values found in the target variable after scaling.')

df['Subject_ID'] = df['Subject_ID'].astype(str)

# Split by subjects to avoid data leakage
unique_subjects = df['Subject_ID'].unique()
train_subjects, test_subjects = train_test_split(unique_subjects, test_size=0.2, random_state=42)

# Apply the split consistently to both inputs and targets
train_indices = df['Subject_ID'].isin(train_subjects).values
test_indices = df['Subject_ID'].isin(test_subjects).values

X_train = X[train_indices].astype(np.float32)
X_test = X[test_indices].astype(np.float32)
y_train = y[train_indices].astype(np.float32)
y_test = y[test_indices].astype(np.float32)

# Verify train-test consistency
print(f"X_train Shape: {X_train.shape}, y_train Shape: {y_train.shape}")
print(f"X_test Shape: {X_test.shape}, y_test Shape: {y_test.shape}")

if len(X_train) != len(y_train) or len(X_test) != len(y_test):
    raise ValueError('Mismatch between X and y sizes after splitting.')

# Final NaN and Inf check
if np.isnan(X_train).any() or np.isnan(y_train).any():
    raise ValueError('NaN detected in training data!')
if np.isnan(X_test).any() or np.isnan(y_test).any():
    raise ValueError('NaN detected in test data!')
if np.isinf(X_train).any() or np.isinf(y_train).any():
    raise ValueError('Infinity detected in training data!')
if np.isinf(X_test).any() or np.isinf(y_test).any():
    raise ValueError('Infinity detected in test data!')
print(f"y_train - Min: {np.min(y_train)}, Max: {np.max(y_train)}, Mean: {np.mean(y_train)}")
print(f"y_test - Min: {np.min(y_test)}, Max: {np.max(y_test)}, Mean: {np.mean(y_test)}")
print(f"y_train - Sample: {y_train[:5]}")
print(f"y_test - Sample: {y_test[:5]}")

if np.isnan(y_train).any() or np.isnan(y_test).any():
    raise ValueError('NaN detected in train or test targets after splitting.')
if np.isinf(y_train).any() or np.isinf(y_test).any():
    raise ValueError('Infinity detected in train or test targets after splitting.')
unique_subjects = df['Subject_ID'].unique()
train_subjects, test_subjects = train_test_split(unique_subjects, test_size=0.2, random_state=42)
train_indices = df['Subject_ID'].isin(train_subjects)
test_indices = df['Subject_ID'].isin(test_subjects)

X_train = X[train_indices]
X_test = X[test_indices]
y_train = y[train_indices]
y_test = y[test_indices]

# 2. Define the custom dataset
class StrokeDataset(Dataset):
    def __init__(self, X, y):
        self.X = X.reshape(-1, 1, len(X_cols)).astype(np.float32)
        self.y = np.array(y, dtype=np.float32)
        if np.isnan(self.y).any() or np.isinf(self.y).any():
            raise ValueError('NaN or infinity detected in target values during dataset initialization.')
        
        if np.isnan(self.X).any() or np.isnan(self.y).any():
            raise ValueError('NaN detected in dataset inputs or targets.')
        if np.isinf(self.X).any() or np.isinf(self.y).any():
            raise ValueError('Infinity detected in dataset inputs or targets.')
        
        if np.isnan(self.X).any() or np.isnan(self.y).any():
            raise ValueError('NaN detected in dataset inputs or targets.')
        if np.isinf(self.X).any() or np.isinf(self.y).any():
            raise ValueError('Infinity detected in dataset inputs or targets.')
        self.y = y

    def __len__(self):
        return len(self.X)

    def __getitem__(self, idx):
        return torch.tensor(self.X[idx], dtype=torch.float32), torch.tensor(self.y[idx], dtype=torch.float32)

train_dataset = StrokeDataset(X_train, y_train)
test_dataset = StrokeDataset(X_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)

# 3. Define the CNN-LSTM model
class CNNLSTM(nn.Module):
    def __init__(self):
        super(CNNLSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv1d(1, 16, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv1d(16, 32, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.Conv1d(32, 64, kernel_size=3, padding=1),
            nn.ReLU()
        )
        self.lstm = nn.LSTM(64 * 22, 64, batch_first=True)
        self.fc = nn.Sequential(
            nn.Linear(64, 32),
            nn.ReLU(),
            nn.Linear(32, 1)
        )

    def forward(self, x):
        x = self.cnn(x)
        x = x.view(x.size(0), 1, -1)  # Flatten to (batch_size, 1, 64 * 22)
        x, _ = self.lstm(x)
        x = x[:, -1, :]  # Take the last time step
        x = self.fc(x)
        return x.squeeze()

# 4. Train the model
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = CNNLSTM().to(device)
optimizer = optim.Adam(model.parameters(), lr=0.0005)
criterion = nn.MSELoss()

num_epochs = 50
train_losses = []
test_losses = []

train_accuracies = []
test_accuracies = []
for epoch in range(num_epochs):
    print(f"Starting Epoch {epoch+1}/{num_epochs}")
    #print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
    #print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
    
    if np.isnan(X_train).any() or np.isnan(y_train).any():
        print("NaN detected in training data!")
        break
    if np.isinf(X_train).any() or np.isinf(y_train).any():
        print("Infinity detected in training data!")
        break
    model.train()
    train_loss = 0.0
    for batch_idx, (X_batch, y_batch) in enumerate(train_loader):
        
        if torch.isnan(X_batch).any() or torch.isnan(y_batch).any():
            raise ValueError('NaN detected in batch inputs or targets.')
        if torch.isinf(X_batch).any() or torch.isinf(y_batch).any():
            raise ValueError('Infinity detected in batch inputs or targets.')
        
        
        if torch.isnan(X_batch).any() or torch.isnan(y_batch).any():
            print("NaN detected in batch!")
            break
        
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        y_pred = model(X_batch)
        loss = criterion(y_pred, y_batch)
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        train_loss += loss.item()
    train_loss /= len(train_loader)
    train_losses.append(train_loss)
    
    # Validation
    correct_train = 0
    total_train = 0
    model.eval()
    with torch.no_grad():
        for X_batch, y_batch in train_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            y_pred = model(X_batch)
            correct_train += torch.sum(torch.abs(y_pred - y_batch) < 0.1).item()
            total_train += y_batch.size(0)
    train_accuracy = correct_train / total_train
    train_accuracies.append(train_accuracy)

    correct_test = 0
    total_test = 0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            y_pred = model(X_batch)
            correct_test += torch.sum(torch.abs(y_pred - y_batch) < 0.1).item()
            total_test += y_batch.size(0)
    test_accuracy = correct_test / total_test
    test_accuracies.append(test_accuracy)

    print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}, Train Accuracy: {train_accuracy:.4f}, Test Accuracy: {test_accuracy:.4f}")
    model.eval()
    test_loss = 0.0
    with torch.no_grad():
        for X_batch, y_batch in test_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            y_pred = model(X_batch)
            loss = criterion(y_pred, y_batch)
            test_loss += loss.item()
    test_loss /= len(test_loader)
    test_losses.append(test_loss)
    print(f"Epoch {epoch+1}/{num_epochs} - Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}")

# Save the trained model
# 5. Evaluate Model Performance
model.eval()
with torch.no_grad():
    y_train_pred = []
    y_train_true = []
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        y_pred = model(X_batch).cpu().numpy()
        y_train_pred.extend(y_pred)
        y_train_true.extend(y_batch.cpu().numpy())

    y_test_pred = []
    y_test_true = []
    for X_batch, y_batch in test_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        y_pred = model(X_batch).cpu().numpy()
        y_test_pred.extend(y_pred)
        y_test_true.extend(y_batch.cpu().numpy())

# Inverse scale the predictions and targets
y_train_pred = scaler_y.inverse_transform(np.array(y_train_pred).reshape(-1, 1)).flatten()
y_train_true = scaler_y.inverse_transform(np.array(y_train_true).reshape(-1, 1)).flatten()
y_test_pred = scaler_y.inverse_transform(np.array(y_test_pred).reshape(-1, 1)).flatten()
y_test_true = scaler_y.inverse_transform(np.array(y_test_true).reshape(-1, 1)).flatten()

# Calculate evaluation metrics
train_mae = mean_absolute_error(y_train_true, y_train_pred)
test_mae = mean_absolute_error(y_test_true, y_test_pred)
train_rmse = np.sqrt(mean_squared_error(y_train_true, y_train_pred))
test_rmse = np.sqrt(mean_squared_error(y_test_true, y_test_pred))
train_r2 = r2_score(y_train_true, y_train_pred)
test_r2 = r2_score(y_test_true, y_test_pred)

print(f"Train MAE: {train_mae:.4f}, Test MAE: {test_mae:.4f}")
print(f"Train RMSE: {train_rmse:.4f}, Test RMSE: {test_rmse:.4f}")
print(f"Train R2: {train_r2:.4f}, Test R2: {test_r2:.4f}")

# Save the trained model and scalers
torch.save(model.state_dict(), 'cnn_lstm_model.pth')
import joblib
joblib.dump(scaler_X, 'scaler_X.pkl')
joblib.dump(scaler_y, 'scaler_y.pkl')

# 5. Plot training and validation loss and accuracy
plt.figure(figsize=(12, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(train_accuracies, label='Train Accuracy')
plt.plot(test_accuracies, label='Test Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
plt.figure(figsize=(12, 6))
plt.plot(train_losses, label='Train Loss')
plt.plot(test_losses, label='Test Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss')
plt.legend()
plt.show()
```
